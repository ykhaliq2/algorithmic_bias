## KNN Clustering Method


### Pre-processing

```python
# Create a copy of the df
df_knn = df.copy()
```

```python
df_knn = df_knn.dropna()
```

```python
# Drop 'derived_race', 'activity_year'
df_knn = df_knn.drop(columns=['derived_race', 'activity_year', 'census_tract'])
```

```python
# Convert 'was_approved' to int64
df_knn['was_approved'] = df_knn['was_approved'].astype('int64')
```

```python
df_knn['covenant_density'] = df_knn['covenant_density'].replace({
    'Low (1-50)': 1,
    'Medium (51-100)': 2,
    'High (101-200)': 3,
    'Very High (>200)': 4
})
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform your features
df_knn_scaled = scaler.fit_transform(df_knn)

df_knn_scaled = pd.DataFrame(df_knn_scaled, columns=df_knn.columns)
```

```python
df_knn_scaled.head()
```

```python
# Create the models

# Import the necessary libraries for k-means and silhouette scores
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples


# Create the libraries
inertias = []
silhouette_scores = []


# Create a for loop to iterate over the different k values
for k in range(1, 11):
    if k == 1:
        inertias.append(0)
        silhouette_scores.append(0)
    else:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(df_knn_scaled)
        inertias.append(kmeans.inertia_)
        silhouette_scores.append(silhouette_score(df_knn_scaled, kmeans.labels_))

# Plot Inertia
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertias, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# Plot Silhouette Scores
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), silhouette_scores, marker='o')
plt.title('Silhouette Score for Different k Values')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()

# Create the silhouette charts
def plot_silhouette(k):
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(df_knn_scaled)
    silhouette_vals = silhouette_samples(df_knn_scaled, cluster_labels)

    plt.figure(figsize=(10, 6))
    y_lower, y_upper = 0, 0

    for i in range(k):
        cluster_silhouette_vals = silhouette_vals[cluster_labels == i]
        cluster_silhouette_vals.sort()
        y_upper += len(cluster_silhouette_vals)

        plt.fill_betweenx(np.arange(y_lower, y_upper),
                         0, cluster_silhouette_vals,
                         alpha=0.7)

        plt.text(-0.05, y_lower + 0.5 * len(cluster_silhouette_vals), f'Cluster {i+1}')
        y_lower = y_upper

    avg_score = silhouette_score(df_knn_scaled, cluster_labels)
    plt.axvline(x=avg_score, color="red", linestyle="--",
                label=f'Average Score: {avg_score:.3f}')

    plt.title(f'Silhouette Analysis for k = {k}')
    plt.xlabel('Silhouette Coefficient')
    plt.ylabel('Cluster')
    plt.legend(loc='lower right')
    plt.show()

# Plot the silhouette scores
for k in [3, 4, 5, 6]:
    plot_silhouette(k)
```

```python
# First Cluster Analysis (k=4)

# Create the variables
kmeans_k1 = KMeans(n_clusters=4, random_state=42)
predict_k1 = kmeans_k1.fit_predict(df_knn_scaled)

# Create the dataframe
re_predict_k1 = pd.DataFrame(predict_k1, columns=['Predict_k1'])

# Print the results
print("Cluster Analysis 1 (k=4) Value Counts:")
print(re_predict_k1['Predict_k1'].value_counts())
print("\n")
```

### Dendrogram

```python
# Create the dendogram

# Import the necessary libraries
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

# Create a linkage matrix
linkage_matrix = linkage(df_knn_scaled, method='complete')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix)
plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

```python
# Create a dataframe for the predictions

# Set the cluster cutoff
cutoff_distance = 20 # Based on the dendogram results
clusters_hc = fcluster(linkage_matrix, cutoff_distance, criterion='distance')

# Create a DataFrame for predictions
re_predict_hc = pd.DataFrame(df_knn_scaled)
re_predict_hc['Predict_hc'] = clusters_hc

# Show value counts for predicted clusters
cluster_counts_hc = re_predict_hc['Predict_hc'].value_counts()
print(cluster_counts_hc)
```

```python
# Create the dendogram

# Create a linkage matrix
linkage_matrix = linkage(df_knn_scaled, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix)
plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

```python
# Create a dataframe for the predictions

# Set the cluster cutoff
cutoff_distance = 90 # Based on the dendogram results
clusters_hw = fcluster(linkage_matrix, cutoff_distance, criterion='distance')

# Create a DataFrame for predictions
re_predict_hw = pd.DataFrame(df_knn_scaled)
re_predict_hw['Predict_hw'] = clusters_hw

# Show value counts for predicted clusters
cluster_counts_hw = re_predict_hw['Predict_hw'].value_counts()
print(cluster_counts_hw)
```
